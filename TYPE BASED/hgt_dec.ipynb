```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import math
import pickle
import random
import numpy as np
import os
from collections import Counter, defaultdict
from sklearn.cluster import KMeans
from sklearn.metrics import normalized_mutual_info_score, adjusted_rand_score, f1_score, accuracy_score
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import dgl
from dgl import heterograph
from dgl.nn import HeteroGraphConv
from rdflib import Graph, URIRef

#############################################
#               MODELE HGT                #
#############################################
class HGTLayer(nn.Module):
    def __init__(self, in_dim, out_dim, ntypes, etypes, n_heads=4, dropout=0.2):
        super().__init__()
        self.in_dim = in_dim
        self.out_dim = out_dim
        self.ntypes = ntypes
        self.etypes = etypes
        self.n_heads = n_heads
        # Assurer que out_dim est divisible par n_heads
        self.d_k = out_dim // n_heads
        self.out_dim = self.d_k * n_heads
        self.dropout = dropout

        # Transformations linéaires par type de nœud
        self.k_linears = nn.ModuleDict({ ntype: nn.Linear(in_dim, self.out_dim) for ntype in ntypes })
        self.q_linears = nn.ModuleDict({ ntype: nn.Linear(in_dim, self.out_dim) for ntype in ntypes })
        self.v_linears = nn.ModuleDict({ ntype: nn.Linear(in_dim, self.out_dim) for ntype in ntypes })
        self.out_linears = nn.ModuleDict({ ntype: nn.Linear(self.out_dim, self.out_dim) for ntype in ntypes })

        # Projections pour les embeddings de relations
        self.relation_proj = nn.ModuleDict({ etype: nn.Linear(768, out_dim) for etype in etypes })

    def forward(self, G, h, relation_embeddings):
        with G.local_scope():
            # Projection des embeddings de relations
            rel_embeddings = {}
            for canonical_etype in G.canonical_etypes:
                src_type, etype, dst_type = canonical_etype
                if etype in self.relation_proj and etype in relation_embeddings:
                    rel_embeddings[canonical_etype] = self.relation_proj[etype](relation_embeddings[etype])

            feat_dict = {}
            for ntype in G.ntypes:
                if G.number_of_nodes(ntype) == 0:
                    continue
                feat_dict[ntype] = torch.zeros((G.number_of_nodes(ntype), self.out_dim),
                                               device=h[ntype].device)
                for canonical_etype in G.canonical_etypes:
                    src_type, etype, dst_type = canonical_etype
                    if src_type == ntype or dst_type == ntype:
                        src_nodes, dst_nodes = G.edges(etype=canonical_etype)
                        if len(src_nodes)==0 or len(dst_nodes)==0:
                            continue
                        q = self.q_linears[dst_type](h[dst_type])
                        k = self.k_linears[src_type](h[src_type])
                        v = self.v_linears[src_type](h[src_type])
                        batch_size = q.size(0)
                        q = q.view(batch_size, self.n_heads, self.d_k)
                        k = k.view(k.size(0), self.n_heads, self.d_k)
                        v = v.view(v.size(0), self.n_heads, self.d_k)
                        q = torch.index_select(q, 0, dst_nodes)
                        k = torch.index_select(k, 0, src_nodes)
                        v = torch.index_select(v, 0, src_nodes)
                        if canonical_etype in rel_embeddings:
                            rel_emb = rel_embeddings[canonical_etype]
                            rel_emb = rel_emb.view(1, self.n_heads, self.d_k)
                            k = k * rel_emb
                            v = v * rel_emb
                        attn = torch.bmm(q, k.transpose(1,2)) / math.sqrt(self.d_k)
                        attn = F.softmax(attn, dim=2)
                        attn = F.dropout(attn, self.dropout, training=self.training)
                        output = torch.bmm(attn, v)
                        output = output.view(-1, self.out_dim)
                        if dst_type == ntype:
                            feat_dict[ntype][dst_nodes] += output
                        else:
                            feat_dict[ntype][src_nodes] += output
            for ntype in G.ntypes:
                if G.number_of_nodes(ntype)==0 or ntype not in feat_dict:
                    continue
                feat_dict[ntype] = self.out_linears[ntype](feat_dict[ntype])
            return feat_dict

class HGT(nn.Module):
    def __init__(self, G, in_dim, hidden_dim, out_dim, n_layers=2, n_heads=4, dropout=0.2):
        super().__init__()
        self.G = G
        self.ntypes = G.ntypes
        self.etypes = G.etypes
        self.in_dim = in_dim
        self.hidden_dim = hidden_dim
        self.out_dim = out_dim
        self.n_layers = n_layers
        self.n_heads = n_heads
        self.dropout = dropout

        self.embeddings = nn.ModuleDict({ ntype: nn.Linear(in_dim, hidden_dim) for ntype in self.ntypes })
        self.layers = nn.ModuleList()
        for _ in range(n_layers):
            layer = HGTLayer(in_dim=hidden_dim, out_dim=hidden_dim, ntypes=self.ntypes, etypes=self.etypes,
                             n_heads=n_heads, dropout=dropout)
            self.layers.append(layer)
        self.outputs = nn.ModuleDict({ ntype: nn.Linear(hidden_dim, out_dim) for ntype in self.ntypes })
        self.batch_norms = nn.ModuleDict({ ntype: nn.BatchNorm1d(out_dim) for ntype in self.ntypes })

    def forward(self, features=None, relation_embeddings=None):
        if features is None:
            h = { ntype: self.embeddings[ntype](self.G.nodes[ntype].data['feat']) for ntype in self.ntypes }
        else:
            h = { ntype: self.embeddings[ntype](features[ntype]) for ntype in self.ntypes }
        for layer in self.layers:
            h_new = layer(self.G, h, relation_embeddings)
            for ntype in h:
                h[ntype] = h_new[ntype] + h[ntype]
        out = {}
        for ntype in self.ntypes:
            out[ntype] = self.outputs[ntype](h[ntype])
            if out[ntype].shape[0] > 1:
                out[ntype] = self.batch_norms[ntype](out[ntype])
        return out

def improved_contrastive_loss(embeddings, graph, temperature=0.07, margin=0.5):
    device = embeddings['author'].device
    pos_pairs = generate_positive_pairs(graph)
    if not pos_pairs:
        raise ValueError("No positive pairs found in the graph")
    h1_list = []
    h2_list = []
    for src_type, src_idx, dst_type, dst_idx in pos_pairs:
        if src_type in embeddings and dst_type in embeddings:
            h1_list.append(embeddings[src_type][src_idx])
            h2_list.append(embeddings[dst_type][dst_idx])
    if not h1_list:
        raise ValueError("No valid embedding pairs found")
    h1 = torch.stack(h1_list)
    h2 = torch.stack(h2_list)
    h1 = F.normalize(h1, p=2, dim=1)
    h2 = F.normalize(h2, p=2, dim=1)
    pos_sim = torch.sum(h1 * h2, dim=1)
    batch_size = h1.size(0)
    neg_sim = torch.mm(h1, h2.t())
    mask = torch.eye(batch_size, device=device)
    neg_sim = neg_sim * (1 - mask)
    k = 3
    hardest_negatives, _ = torch.topk(neg_sim, k=min(k, batch_size-1), dim=1)
    hardest_negative_mean = torch.mean(hardest_negatives, dim=1)
    pos_term = -torch.log(torch.exp(pos_sim / temperature) / (torch.exp(pos_sim / temperature) + torch.exp(hardest_negative_mean / temperature) + 1e-8))
    margin_term = torch.clamp(margin - pos_sim, min=0)
    loss = pos_term.mean() + 0.9 * margin_term.mean()
    return loss

def generate_positive_pairs(g):
    pos_pairs = []
    key_relations = [('publication', 'creator', 'author'), ('author', 'hasDomain', 'domain'),
                     ('publication', 'isPartOf', 'conference'), ('conference', 'publishesDomain', 'domain')]
    inverse_relations = [('author', 'created', 'publication'), ('domain', 'hasAuthor', 'author'),
                         ('conference', 'hasPublication', 'publication'), ('domain', 'isPublishedIn', 'conference')]
    second_order_relations = []
    authors_by_domain = {}
    if ('author', 'hasDomain', 'domain') in g.canonical_etypes:
        src, dst = g.edges(etype=('author', 'hasDomain', 'domain'))
        for s, d in zip(src.tolist(), dst.tolist()):
            authors_by_domain.setdefault(d, []).append(s)
    for domain, authors in authors_by_domain.items():
        if len(authors) > 1:
            if len(authors) > 30:
                authors = random.sample(authors, 30)
            for i in range(len(authors)):
                for j in range(i+1, len(authors)):
                    second_order_relations.append(('author', authors[i], 'author', authors[j]))
    all_relations = key_relations + inverse_relations
    relation_pairs = {}
    for src_type, rel_type, dst_type in all_relations:
        try:
            if (src_type, rel_type, dst_type) in g.canonical_etypes:
                src, dst = g.edges(etype=(src_type, rel_type, dst_type))
                if len(src)>0 and len(dst)>0:
                    max_src = g.number_of_nodes(src_type) - 1
                    max_dst = g.number_of_nodes(dst_type) - 1
                    valid_pairs = []
                    for s, d in zip(src.tolist(), dst.tolist()):
                        if s <= max_src and d <= max_dst:
                            valid_pairs.append((src_type, s, dst_type, d))
                    relation_pairs[(src_type, rel_type, dst_type)] = valid_pairs
        except Exception as e:
            print(f"Warning: Skipping {src_type}-{rel_type}-{dst_type}: {e}")
            continue
    max_pairs_per_relation = 800
    for rel, pairs in relation_pairs.items():
        max_pairs = max_pairs_per_relation * 5 if rel[0]=='author' and rel[2]=='domain' else max_pairs_per_relation
        sampled_pairs = random.sample(pairs, max_pairs) if len(pairs) > max_pairs else pairs
        pos_pairs.extend(sampled_pairs)
    max_second_order = 1000
    if len(second_order_relations) > max_second_order:
        second_order_relations = random.sample(second_order_relations, max_second_order)
    pos_pairs.extend(second_order_relations)
    random.shuffle(pos_pairs)
    return pos_pairs

def train_hgt(g, node_features, relation_embeddings, num_epochs=100, lr=1e-6):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    relation_embeddings = { k: v.to(device) for k, v in relation_embeddings.items() }
    torch.manual_seed(42)
    model = HGT(G=g, in_dim=768, hidden_dim=768, out_dim=384, n_layers=2, n_heads=12, dropout=0.05).to(device)
    optimizer = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=0.00005, betas=(0.9,0.999))
    scheduler = torch.optim.lr_scheduler.OneCycleLR(optimizer, max_lr=lr*5, steps_per_epoch=1, epochs=num_epochs, pct_start=0.05, anneal_strategy='cos')
    best_loss = float('inf')
    patience = 70
    patience_counter = 0
    best_model_state = None
    ema_loss = None
    alpha = 0.8
    print(f"Training on {device} for {num_epochs} epochs...")
    for epoch in range(num_epochs):
        model.train()
        optimizer.zero_grad()
        embeddings = model(features=None, relation_embeddings=relation_embeddings)
        for ntype in embeddings:
            norm = torch.norm(embeddings[ntype], p=2, dim=1, keepdim=True)
            embeddings[ntype] = embeddings[ntype] / (norm + 1e-8)
        loss = improved_contrastive_loss(embeddings, g, temperature=0.07, margin=0.5)
        ema_loss = loss.item() if ema_loss is None else alpha*ema_loss + (1-alpha)*loss.item()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        loss.backward()
        optimizer.step()
        scheduler.step()
        if ema_loss < best_loss:
            best_loss = ema_loss
            patience_counter = 0
            best_model_state = model.state_dict().copy()
        else:
            patience_counter += 1
        if patience_counter >= patience:
            print(f"Early stopping at epoch {epoch}")
            model.load_state_dict(best_model_state)
            break
        if (epoch+1) % 10 == 0:
            print(f"Epoch {epoch+1}/{num_epochs}, Loss: {ema_loss:.4f}")
            print(f"Learning rate: {optimizer.param_groups[0]['lr']:.6f}")
    if best_model_state is not None:
        model.load_state_dict(best_model_state)
    model.eval()
    with torch.no_grad():
        embeddings = model(features=None, relation_embeddings=relation_embeddings)
    return model, embeddings

#############################################
#          MODULE JOINT DEC               #
#############################################
class JointDEC(nn.Module):
    """
    Joint Deep Embedded Clustering module pour plusieurs types de nœuds.
    """
    def __init__(self, embeddings_dim, n_clusters_dict, alpha=1.0):
        super(JointDEC, self).__init__()
        self.embeddings_dim = embeddings_dim
        self.n_clusters_dict = n_clusters_dict
        self.alpha = alpha
        self.cluster_centers = nn.ParameterDict({
            ntype: nn.Parameter(torch.zeros(n_clusters, embeddings_dim))
            for ntype, n_clusters in n_clusters_dict.items()
        })
        self.transform_layers = nn.ModuleDict({
            ntype: nn.Sequential(
                nn.Linear(embeddings_dim, embeddings_dim),
                nn.BatchNorm1d(embeddings_dim),
                nn.ReLU(),
                nn.Linear(embeddings_dim, embeddings_dim)
            )
            for ntype in n_clusters_dict.keys()
        })
    def forward(self, embeddings_dict):
        transformed_embeddings = {}
        q_dict = {}
        for ntype, embeddings in embeddings_dict.items():
            if ntype in self.transform_layers and ntype in self.cluster_centers:
                transformed = self.transform_layers[ntype](embeddings)
                transformed = F.normalize(transformed, p=2, dim=1)
                transformed_embeddings[ntype] = transformed
                q = self._soft_assignment(transformed, self.cluster_centers[ntype])
                q_dict[ntype] = q
        return transformed_embeddings, q_dict
    def _soft_assignment(self, embeddings, cluster_centers):
        q = 1.0 / (1.0 + torch.sum((embeddings.unsqueeze(1) - cluster_centers.unsqueeze(0))**2, dim=2) / self.alpha)
        q = q ** ((self.alpha+1.0)/2.0)
        q = q / torch.sum(q, dim=1, keepdim=True)
        return q
    def target_distribution(self, q):
        weight = q ** 2 / q.sum(0)
        return (weight.t() / weight.sum(1)).t()
    def init_cluster_centers(self, embeddings_dict):
        for ntype, embeddings in embeddings_dict.items():
            if ntype in self.n_clusters_dict:
                n_clusters = self.n_clusters_dict[ntype]
                if ntype in self.transform_layers:
                    with torch.no_grad():
                        transformed = self.transform_layers[ntype](embeddings)
                        transformed = F.normalize(transformed, p=2, dim=1)
                else:
                    transformed = embeddings
                kmeans = KMeans(n_clusters=n_clusters, n_init=20, random_state=42)
                embeddings_np = transformed.detach().cpu().numpy()
                kmeans.fit(embeddings_np)
                cluster_centers = torch.tensor(kmeans.cluster_centers_, dtype=torch.float32).to(embeddings.device)
                self.cluster_centers[ntype].data = cluster_centers
                print(f"Initialized {ntype} cluster centers using KMeans")

def train_joint_dec(embeddings_dict, n_clusters_dict, true_labels_dict=None,
                    max_iters=100, update_interval=10, tol=1e-3, device='cuda'):
    embeddings_dict = { k: v.to(device) if isinstance(v, torch.Tensor) else torch.tensor(v, dtype=torch.float32).to(device)
                        for k, v in embeddings_dict.items() }
    embeddings_dim = next(iter(embeddings_dict.values())).shape[1]
    model = JointDEC(embeddings_dim, n_clusters_dict, alpha=1.0).to(device)
    model.init_cluster_centers(embeddings_dict)
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
    best_loss = float('inf')
    best_state = None
    loss_history = []
    patience = 10
    patience_counter = 0

    for iter_num in range(max_iters):
        transformed_embeddings, q_dict = model(embeddings_dict)
        # Mise à jour périodique de la target distribution avec détachement (detach)
        if iter_num % update_interval == 0:
            p_dict = { ntype: model.target_distribution(q).detach() for ntype, q in q_dict.items() }
            if true_labels_dict is not None:
                print(f"\nIteration {iter_num}/{max_iters}:")
                for ntype, q in q_dict.items():
                    if ntype in true_labels_dict and true_labels_dict[ntype] is not None:
                        _, y_pred = torch.max(q.data, 1)
                        y_pred = y_pred.cpu().numpy()
                        y_true = true_labels_dict[ntype]
                        nmi = normalized_mutual_info_score(y_true, y_pred)
                        ari = adjusted_rand_score(y_true, y_pred)
                        f1, acc = calculate_clustering_f1_acc(y_true, y_pred)
                        print(f"{ntype} - NMI: {nmi:.4f}, ARI: {ari:.4f}, F1: {f1:.4f}, ACC: {acc:.4f}")
        loss = 0
        for ntype, q in q_dict.items():
            if ntype in p_dict:
                loss += F.kl_div(q.log(), p_dict[ntype], reduction='batchmean')
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        if (iter_num+1) % 10 == 0:
            print(f"Iteration {iter_num+1}/{max_iters}, Loss: {loss.item():.4f}")
            loss_history.append(loss.item())
            if loss < best_loss:
                best_loss = loss
                best_state = model.state_dict().copy()
                patience_counter = 0
            else:
                patience_counter += 1
            if patience_counter >= patience:
                print(f"Early stopping at iteration {iter_num+1}")
                model.load_state_dict(best_state)
                break
            if len(loss_history)>1 and abs(loss_history[-1]-loss_history[-2]) < tol:
                print(f"Converged at iteration {iter_num+1}")
                break
    model.eval()
    with torch.no_grad():
        _, q_dict = model(embeddings_dict)
        cluster_assignments = {}
        for ntype, q in q_dict.items():
            _, assignments = torch.max(q.data, 1)
            cluster_assignments[ntype] = assignments.cpu().numpy()
    return model, cluster_assignments

def calculate_clustering_f1_acc(true_labels, pred_labels):
    """
    Calculate F1 and Accuracy for clustering using optimal cluster-to-class assignment.
    """
    from scipy.optimize import linear_sum_assignment

    unique_true = np.unique(true_labels)
    unique_pred = np.unique(pred_labels)

    # Create contingency matrix
    contingency = np.zeros((len(unique_pred), len(unique_true)), dtype=int)
    for i, pred_label in enumerate(unique_pred):
        for j, true_label in enumerate(unique_true):
            contingency[i, j] = np.sum((pred_labels == pred_label) & (true_labels == true_label))

    # Use Hungarian algorithm to find optimal assignment
    row_indices, col_indices = linear_sum_assignment(-contingency)

    # Create optimal mapping
    optimal_mapping = {}
    for i, j in zip(row_indices, col_indices):
        if i < len(unique_pred) and j < len(unique_true):
            optimal_mapping[unique_pred[i]] = unique_true[j]

    # Apply optimal mapping to predicted labels
    aligned_pred_labels = np.array([optimal_mapping.get(pred, pred) for pred in pred_labels])

    # Calculate accuracy (percentage of correctly aligned nodes)
    accuracy = np.sum(aligned_pred_labels == true_labels) / len(true_labels)

    # Calculate F1 score (macro average)
    f1 = f1_score(true_labels, aligned_pred_labels, average='macro', zero_division=0)

    return f1, accuracy

def evaluate_joint_dec_clustering(cluster_assignments_dict, true_labels_dict):
    results_dict = {}
    for ntype, pred_labels in cluster_assignments_dict.items():
        if ntype in true_labels_dict and true_labels_dict[ntype] is not None:
            true_labels = true_labels_dict[ntype]
            if len(true_labels)!=len(pred_labels):
                print(f"Warning: Length mismatch for {ntype}. True: {len(true_labels)}, Pred: {len(pred_labels)}")
                continue
            nmi = normalized_mutual_info_score(true_labels, pred_labels)
            ari = adjusted_rand_score(true_labels, pred_labels)
            f1, acc = calculate_clustering_f1_acc(true_labels, pred_labels)
            results_dict[ntype] = {'nmi': nmi, 'ari': ari, 'f1': f1, 'acc': acc}
            print(f"{ntype} Clustering Evaluation:")
            print(f"NMI: {nmi:.4f}")
            print(f"ARI: {ari:.4f}")
            print(f"F1: {f1:.4f}")
            print(f"ACC: {acc:.4f}")
    return results_dict

def save_clustering_results(results_dict, cluster_assignments_dict, true_labels_dict, output_dir="Hgt with DEC"):
    """
    Save clustering results to clustering_results_summary.txt file with enhanced alignment information
    """
    import datetime

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    results_file = os.path.join(output_dir, "clustering_results_summary.txt")

    with open(results_file, 'w', encoding='utf-8') as f:
        f.write("="*80 + "\n")
        f.write("HGT + DEC CLUSTERING RESULTS SUMMARY\n")
        f.write("="*80 + "\n")
        f.write(f"Generated on: {datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
        f.write(f"Method: Heterogeneous Graph Transformer (HGT) + Deep Embedded Clustering (DEC)\n")
        f.write(f"Enhanced with: Color-Aligned Visualizations and Alignment Quality Assessment\n")
        f.write("="*80 + "\n\n")

        # Overall summary
        f.write("OVERALL CLUSTERING PERFORMANCE:\n")
        f.write("-" * 40 + "\n")

        total_nmi = 0
        total_ari = 0
        total_f1 = 0
        total_acc = 0
        valid_types = 0

        total_alignment_quality = 0

        for ntype, metrics in results_dict.items():
            if 'nmi' in metrics and 'ari' in metrics:
                f.write(f"{ntype.upper()} CLUSTERING:\n")
                f.write(f"  - NMI (Normalized Mutual Information): {metrics['nmi']:.4f}\n")
                f.write(f"  - ARI (Adjusted Rand Index): {metrics['ari']:.4f}\n")
                if 'f1' in metrics:
                    f.write(f"  - F1 Score: {metrics['f1']:.4f}\n")
                if 'acc' in metrics:
                    f.write(f"  - Accuracy: {metrics['acc']:.4f}\n")

                if ntype in cluster_assignments_dict and ntype in true_labels_dict:
                    n_true_clusters = len(np.unique(true_labels_dict[ntype]))
                    n_pred_clusters = len(np.unique(cluster_assignments_dict[ntype]))
                    n_nodes = len(cluster_assignments_dict[ntype])

                    f.write(f"  - Number of nodes: {n_nodes}\n")
                    f.write(f"  - True clusters: {n_true_clusters}\n")
                    f.write(f"  - Predicted clusters: {n_pred_clusters}\n")

                    # Calculate and include alignment quality
                    alignment_mapping = create_cluster_alignment_mapping(
                        true_labels_dict[ntype], cluster_assignments_dict[ntype])
                    alignment_quality = calculate_alignment_quality(
                        true_labels_dict[ntype], cluster_assignments_dict[ntype], alignment_mapping)
                    f.write(f"  - Alignment Quality: {alignment_quality:.4f}\n")
                    f.write(f"  - Optimal Alignments: {len(alignment_mapping)}\n")
                    total_alignment_quality += alignment_quality

                f.write("\n")
                total_nmi += metrics['nmi']
                total_ari += metrics['ari']
                if 'f1' in metrics:
                    total_f1 += metrics['f1']
                if 'acc' in metrics:
                    total_acc += metrics['acc']
                valid_types += 1

        if valid_types > 0:
            avg_nmi = total_nmi / valid_types
            avg_ari = total_ari / valid_types
            avg_f1 = total_f1 / valid_types
            avg_acc = total_acc / valid_types
            avg_alignment_quality = total_alignment_quality / valid_types
            f.write(f"AVERAGE PERFORMANCE ACROSS ALL NODE TYPES:\n")
            f.write(f"  - Average NMI: {avg_nmi:.4f}\n")
            f.write(f"  - Average ARI: {avg_ari:.4f}\n")
            f.write(f"  - Average F1: {avg_f1:.4f}\n")
            f.write(f"  - Average ACC: {avg_acc:.4f}\n")
            f.write(f"  - Average Alignment Quality: {avg_alignment_quality:.4f}\n")
            f.write(f"  🎨 Enhanced with color-aligned visualizations\n\n")

        # Detailed cluster analysis
        f.write("DETAILED CLUSTER ANALYSIS:\n")
        f.write("-" * 40 + "\n")

        for ntype in results_dict.keys():
            if ntype in cluster_assignments_dict and ntype in true_labels_dict:
                f.write(f"\n{ntype.upper()} CLUSTER DISTRIBUTION:\n")

                pred_labels = cluster_assignments_dict[ntype]
                true_labels = true_labels_dict[ntype]

                # Cluster size distribution
                unique_pred, counts_pred = np.unique(pred_labels, return_counts=True)
                f.write(f"  Predicted cluster sizes: {dict(zip(unique_pred, counts_pred))}\n")

                unique_true, counts_true = np.unique(true_labels, return_counts=True)
                f.write(f"  True cluster sizes: {dict(zip(unique_true, counts_true))}\n")

                # Cluster purity analysis
                total_correct = 0
                total_nodes = len(pred_labels)

                f.write(f"  Cluster purity analysis:\n")
                for cluster in unique_pred:
                    cluster_indices = np.where(pred_labels == cluster)[0]
                    cluster_true_labels = true_labels[cluster_indices]
                    unique_true_in_cluster, counts = np.unique(cluster_true_labels, return_counts=True)
                    dominant_label = unique_true_in_cluster[np.argmax(counts)]
                    dominant_count = np.max(counts)
                    total_correct += dominant_count
                    purity = dominant_count / len(cluster_indices)

                    f.write(f"    Cluster {cluster}: {len(cluster_indices)} nodes, ")
                    f.write(f"dominant class {dominant_label} ({dominant_count} nodes, {purity:.2%})\n")

                overall_purity = total_correct / total_nodes
                f.write(f"  Overall purity for {ntype}: {overall_purity:.4f}\n")

        # Model configuration
        f.write("\n" + "="*80 + "\n")
        f.write("MODEL CONFIGURATION:\n")
        f.write("="*80 + "\n")
        f.write("HGT Model Parameters:\n")
        f.write("  - Input dimension: 768\n")
        f.write("  - Hidden dimension: 768\n")
        f.write("  - Output dimension: 384\n")
        f.write("  - Number of layers: 2\n")
        f.write("  - Number of attention heads: 12\n")
        f.write("  - Dropout rate: 0.05\n")
        f.write("  - Learning rate: 1e-6 (with OneCycleLR scheduler)\n")
        f.write("  - Weight decay: 0.00005\n")
        f.write("  - Early stopping patience: 70 epochs\n\n")

        f.write("DEC Model Parameters:\n")
        f.write("  - Alpha parameter: 1.0\n")
        f.write("  - Maximum iterations: 20\n")
        f.write("  - Update interval: 10\n")
        f.write("  - Learning rate: 0.001\n")
        f.write("  - Early stopping patience: 10 iterations\n\n")

        f.write("Loss Functions:\n")
        f.write("  - HGT: Improved contrastive loss with hard negative mining\n")
        f.write("  - DEC: KL divergence between soft assignments and target distribution\n")

        f.write("\n" + "="*80 + "\n")
        f.write("END OF REPORT\n")
        f.write("="*80 + "\n")

    print(f"\n✓ Clustering results saved to: {results_file}")
    return results_file

def evaluate_clustering_with_dec(embeddings, true_labels, n_clusters):
    """
    Evaluate clustering performance using HGT+DEC with enhanced alignment quality assessment
    """
    if isinstance(embeddings, torch.Tensor):
        embeddings_tensor = embeddings
    else:
        embeddings_tensor = torch.tensor(embeddings, dtype=torch.float32)

    embeddings_tensor = F.normalize(embeddings_tensor, p=2, dim=1)
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    embeddings_tensor = embeddings_tensor.to(device)

    print(f"🔄 Running HGT+DEC clustering with {n_clusters} clusters...")

    embeddings_dict = {'current': embeddings_tensor}
    n_clusters_dict = {'current': n_clusters}
    true_labels_dict = {'current': true_labels}

    model, cluster_assignments = train_joint_dec(embeddings_dict, n_clusters_dict, true_labels_dict,
                                                  max_iters=100, update_interval=10, device=device)

    pred_labels = cluster_assignments['current']

    # Calculate traditional metrics
    nmi = normalized_mutual_info_score(true_labels, pred_labels)
    ari = adjusted_rand_score(true_labels, pred_labels)
    f1, acc = calculate_clustering_f1_acc(true_labels, pred_labels)

    # Calculate alignment quality
    alignment_mapping = create_cluster_alignment_mapping(true_labels, pred_labels)
    alignment_quality = calculate_alignment_quality(true_labels, pred_labels, alignment_mapping)

    print(f"✅ HGT+DEC clustering completed: NMI={nmi:.4f}, ARI={ari:.4f}, F1={f1:.4f}, ACC={acc:.4f}, Alignment Quality={alignment_quality:.4f}")

    return nmi, ari, f1, acc, pred_labels

def calculate_alignment_quality(true_labels, pred_labels, alignment_mapping):
    """
    Calculate the quality of cluster-domain alignment.

    Args:
        true_labels: Array of true domain labels
        pred_labels: Array of predicted cluster labels
        alignment_mapping: Mapping from predicted clusters to true domains

    Returns:
        float: Alignment quality score (0-1, higher is better)
    """
    if not alignment_mapping:
        return 0.0

    total_aligned = 0
    total_nodes = len(true_labels)

    for cluster, domain in alignment_mapping.items():
        # Count nodes that are correctly aligned
        correctly_aligned = np.sum((pred_labels == cluster) & (true_labels == domain))
        total_aligned += correctly_aligned

    return total_aligned / total_nodes if total_nodes > 0 else 0.0

def create_cluster_alignment_mapping(true_labels, pred_labels):
    """
    Create a mapping between predicted clusters and true domains based on maximum overlap.

    Args:
        true_labels: Array of true domain labels
        pred_labels: Array of predicted cluster labels

    Returns:
        dict: Mapping from predicted cluster labels to corresponding true domain labels
    """
    from scipy.optimize import linear_sum_assignment

    unique_true = np.unique(true_labels)
    unique_pred = np.unique(pred_labels)

    # Create contingency matrix
    contingency = np.zeros((len(unique_pred), len(unique_true)), dtype=int)
    for i, pred_label in enumerate(unique_pred):
        for j, true_label in enumerate(unique_true):
            contingency[i, j] = np.sum((pred_labels == pred_label) & (true_labels == true_label))

    # Use Hungarian algorithm to find optimal assignment
    row_indices, col_indices = linear_sum_assignment(-contingency)

    # Create mapping
    alignment_mapping = {}
    for i, j in zip(row_indices, col_indices):
        if i < len(unique_pred) and j < len(unique_true):
            alignment_mapping[unique_pred[i]] = unique_true[j]

    return alignment_mapping

def create_aligned_color_scheme(unique_true, unique_pred, alignment_mapping):
    """
    Create an aligned color scheme where corresponding clusters and domains use the same colors.

    Args:
        unique_true: Array of unique true domain labels
        unique_pred: Array of unique predicted cluster labels
        alignment_mapping: Mapping from predicted clusters to true domains

    Returns:
        dict: Color scheme with 'true' and 'pred' keys containing color mappings
    """
    # Use a high-quality color palette with enough distinct colors
    import matplotlib.colors as mcolors

    # Create a comprehensive color palette
    base_colors = [
        '#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd',
        '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf',
        '#aec7e8', '#ffbb78', '#98df8a', '#ff9896', '#c5b0d5',
        '#c49c94', '#f7b6d3', '#c7c7c7', '#dbdb8d', '#9edae5'
    ]

    # Extend color palette if needed
    if len(unique_true) > len(base_colors):
        # Generate additional colors using HSV space
        additional_colors = []
        for i in range(len(unique_true) - len(base_colors)):
            hue = (i * 0.618033988749895) % 1  # Golden ratio for good distribution
            color = mcolors.hsv_to_rgb([hue, 0.7, 0.9])
            additional_colors.append(mcolors.to_hex(color))
        base_colors.extend(additional_colors)

    # Assign colors to true domains
    true_colors = {}
    for i, domain in enumerate(unique_true):
        true_colors[domain] = base_colors[i % len(base_colors)]

    # Assign colors to predicted clusters based on alignment
    pred_colors = {}
    unaligned_color_idx = len(unique_true)  # Start after true domain colors

    for cluster in unique_pred:
        if cluster in alignment_mapping:
            # Use the same color as the corresponding true domain
            corresponding_domain = alignment_mapping[cluster]
            pred_colors[cluster] = true_colors[corresponding_domain]
        else:
            # Use a distinct color for unaligned clusters
            pred_colors[cluster] = base_colors[unaligned_color_idx % len(base_colors)]
            unaligned_color_idx += 1

    return {
        'true': true_colors,
        'pred': pred_colors
    }

#############################################
#           VISUALISATION UTILES           #
#############################################
def visualize_clusters(embeddings, true_labels, pred_labels, title="Cluster Visualization"):
    """
    Visualize clusters using t-SNE with enhanced color alignment and display.
    Features enhanced color alignment where corresponding clusters use the same colors.
    """
    # Create output directory if it doesn't exist
    output_dir = "Hgt with DEC"
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    if isinstance(embeddings, torch.Tensor):
        embeddings = embeddings.detach().cpu().numpy()

    # Apply t-SNE with better parameters
    tsne = TSNE(n_components=2, perplexity=min(30, len(embeddings)-1),
                random_state=42, n_iter=1000, learning_rate=200)
    embeddings_2d = tsne.fit_transform(embeddings)

    # Create figure with improved layout
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

    # Get unique labels
    unique_true = np.unique(true_labels)
    unique_pred = np.unique(pred_labels)

    # Create alignment mapping between predicted clusters and true domains
    alignment_mapping = create_cluster_alignment_mapping(true_labels, pred_labels)

    # Create aligned color scheme
    aligned_colors = create_aligned_color_scheme(unique_true, unique_pred, alignment_mapping)

    # Plot true labels with aligned colors
    for i, label in enumerate(unique_true):
        mask = true_labels == label
        color = aligned_colors['true'][label]
        ax1.scatter(embeddings_2d[mask, 0], embeddings_2d[mask, 1],
                   c=[color], label=f'Domain {int(label)}',
                   alpha=0.7, s=60, edgecolors='black', linewidths=0.3)

    ax1.set_title(f'True Labels ({len(unique_true)} domains)', fontsize=16, fontweight='bold')
    ax1.set_xlabel('t-SNE Dimension 1', fontsize=14)
    ax1.set_ylabel('t-SNE Dimension 2', fontsize=14)
    ax1.grid(True, alpha=0.3)

    # Improved legend positioning and styling
    legend1 = ax1.legend(bbox_to_anchor=(1.05, 1), loc='upper left',
                        title="True Domains", title_fontsize=12, fontsize=10)
    legend1.get_frame().set_facecolor('white')
    legend1.get_frame().set_alpha(0.9)

    # Plot predicted clusters with aligned colors
    for i, label in enumerate(unique_pred):
        mask = pred_labels == label
        color = aligned_colors['pred'][label]
        # Find the corresponding true domain for this cluster
        corresponding_domain = alignment_mapping.get(label, None)
        cluster_label = f'Cluster {int(label)}'
        if corresponding_domain is not None:
            cluster_label += f' → Domain {int(corresponding_domain)}'

        ax2.scatter(embeddings_2d[mask, 0], embeddings_2d[mask, 1],
                   c=[color], label=cluster_label,
                   alpha=0.7, s=60, edgecolors='black', linewidths=0.3)

    ax2.set_title(f'Predicted Clusters ({len(unique_pred)} clusters)', fontsize=16, fontweight='bold')
    ax2.set_xlabel('t-SNE Dimension 1', fontsize=14)
    ax2.set_ylabel('t-SNE Dimension 2', fontsize=14)
    ax2.grid(True, alpha=0.3)

    # Improved legend positioning and styling
    legend2 = ax2.legend(bbox_to_anchor=(1.05, 1), loc='upper left',
                        title="Predicted Clusters", title_fontsize=12, fontsize=10)
    legend2.get_frame().set_facecolor('white')
    legend2.get_frame().set_alpha(0.9)

    # Add cluster alignment analysis with color coordination info
    alignment_info = f"🎨 Color-Aligned Visualization: Matching clusters and domains use the same colors"
    plt.suptitle(f'{title}\nHGT+DEC Cluster-Domain Alignment Analysis\n{alignment_info}',
                fontsize=16, fontweight='bold', y=0.95)

    # Adjust layout to accommodate legends and title
    plt.tight_layout()
    plt.subplots_adjust(right=0.85, top=0.85)

    # Save in the specified directory
    filename = f'{title.lower().replace(" ", "_").replace("(", "").replace(")", "")}.png'
    filepath = os.path.join(output_dir, filename)
    plt.savefig(filepath, dpi=300, bbox_inches='tight')
    print(f"📸 Visualization saved to: {filepath}")
    plt.show()

    # Print enhanced cluster alignment information
    print(f"\n=== {title} - Enhanced HGT+DEC Cluster Alignment Analysis ===")
    print(f"Number of true domains: {len(unique_true)}")
    print(f"Number of predicted clusters: {len(unique_pred)}")
    print(f"🎨 Visual Enhancement: Corresponding clusters and domains use matching colors")

    # Print alignment mapping information
    print(f"\n🔗 Cluster-Domain Alignment Mapping:")
    for cluster, domain in alignment_mapping.items():
        cluster_size = np.sum(pred_labels == cluster)
        domain_size = np.sum(true_labels == domain)
        overlap = np.sum((pred_labels == cluster) & (true_labels == domain))
        precision = (overlap / cluster_size) * 100 if cluster_size > 0 else 0
        recall = (overlap / domain_size) * 100 if domain_size > 0 else 0
        print(f"   Cluster {int(cluster)} ↔ Domain {int(domain)}: "
              f"{overlap} nodes (Precision: {precision:.1f}%, Recall: {recall:.1f}%)")

    # Calculate and display alignment quality metrics
    total_aligned = sum(np.sum((pred_labels == cluster) & (true_labels == domain))
                       for cluster, domain in alignment_mapping.items())
    total_nodes = len(true_labels)
    alignment_accuracy = (total_aligned / total_nodes) * 100 if total_nodes > 0 else 0

    # Calculate F1 and ACC metrics
    f1, acc = calculate_clustering_f1_acc(true_labels, pred_labels)

    print(f"\n✅ HGT+DEC Clustering Performance Metrics:")
    print(f"   F1 Score: {f1:.4f}")
    print(f"   Accuracy: {acc:.4f}")
    print(f"   Total nodes correctly aligned: {total_aligned}/{total_nodes} ({alignment_accuracy:.1f}%)")
    print(f"   Number of optimal alignments: {len(alignment_mapping)}")

    # Generate and save alignment report
    save_alignment_report(true_labels, pred_labels, alignment_mapping, title, output_dir)

def save_alignment_report(true_labels, pred_labels, alignment_mapping, title, output_dir):
    """
    Save a detailed alignment report to a text file for HGT+DEC clustering.

    Args:
        true_labels: Array of true domain labels
        pred_labels: Array of predicted cluster labels
        alignment_mapping: Mapping from predicted clusters to true domains
        title: Title for the report
        output_dir: Directory to save the report
    """
    from datetime import datetime

    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    report_content = f"""======================================================================
HGT+DEC CLUSTER-DOMAIN ALIGNMENT REPORT
======================================================================
Generated on: {timestamp}
Analysis: {title}
Clustering Method: Heterogeneous Graph Transformer + Deep Embedded Clustering

======================================================================
ALIGNMENT SUMMARY
======================================================================

Total nodes analyzed: {len(true_labels):,}
Number of true domains: {len(np.unique(true_labels))}
Number of predicted clusters: {len(np.unique(pred_labels))}
Number of optimal alignments: {len(alignment_mapping)}

======================================================================
DETAILED ALIGNMENT MAPPING
======================================================================

"""

    # Calculate detailed metrics for each alignment
    for cluster, domain in alignment_mapping.items():
        cluster_size = np.sum(pred_labels == cluster)
        domain_size = np.sum(true_labels == domain)
        overlap = np.sum((pred_labels == cluster) & (true_labels == domain))
        precision = (overlap / cluster_size) * 100 if cluster_size > 0 else 0
        recall = (overlap / domain_size) * 100 if domain_size > 0 else 0
        f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

        report_content += f"""Cluster {int(cluster)} ↔ Domain {int(domain)}:
  - Overlap: {overlap} nodes
  - Cluster size: {cluster_size} nodes
  - Domain size: {domain_size} nodes
  - Precision: {precision:.2f}%
  - Recall: {recall:.2f}%
  - F1-Score: {f1:.2f}%

"""

    # Calculate overall alignment quality
    total_aligned = sum(np.sum((pred_labels == cluster) & (true_labels == domain))
                       for cluster, domain in alignment_mapping.items())
    alignment_accuracy = (total_aligned / len(true_labels)) * 100

    report_content += f"""======================================================================
OVERALL ALIGNMENT QUALITY
======================================================================

Total correctly aligned nodes: {total_aligned}/{len(true_labels)}
Overall alignment accuracy: {alignment_accuracy:.2f}%

======================================================================
HGT+DEC CLUSTERING ADVANTAGES
======================================================================

✅ Joint Learning: HGT learns graph representations while DEC performs
   clustering simultaneously, leading to better cluster-aware embeddings.

✅ Deep Embedding: DEC transforms embeddings through neural networks
   before clustering, capturing non-linear patterns.

✅ Iterative Refinement: DEC iteratively refines cluster assignments
   using target distribution updates.

✅ Multi-Node Type Support: Joint clustering across different node types
   (authors, publications, venues) with shared representation learning.

✅ Relation-Aware: HGT incorporates relation embeddings for better
   heterogeneous graph understanding.

======================================================================
VISUALIZATION ENHANCEMENT
======================================================================

🎨 Color Alignment: The visualization uses matching colors for corresponding
   clusters and domains, making it easier to visually assess alignment quality.

✅ Optimal alignments are clearly visible through color coordination
🔍 Deep embedding structure provides additional cluster separability

======================================================================
"""

    # Save the report
    report_filename = f"{title.lower().replace(' ', '_').replace('(', '').replace(')', '')}_hgt_dec_alignment_report.txt"
    report_filepath = os.path.join(output_dir, report_filename)

    with open(report_filepath, 'w', encoding='utf-8') as f:
        f.write(report_content)

    print(f"📄 HGT+DEC alignment report saved to: {report_filepath}")
    return report_filepath

def analyze_clusters(embeddings, true_labels, pred_labels, id_node_map):
    if isinstance(embeddings, torch.Tensor):
        embeddings = embeddings.detach().cpu().numpy()
    unique_clusters = np.unique(pred_labels)
    print("\n" + "="*50)
    print("CLUSTER ANALYSIS REPORT")
    print("="*50)
    total_correct = 0
    total_nodes = len(pred_labels)
    for cluster in unique_clusters:
        cluster_indices = np.where(pred_labels == cluster)[0]
        cluster_true_labels = true_labels[cluster_indices]
        unique_true_labels, counts = np.unique(cluster_true_labels, return_counts=True)
        dominant_label = unique_true_labels[np.argmax(counts)]
        dominant_count = np.max(counts)
        total_correct += dominant_count
        print(f"\nCluster {cluster}:")
        print(f"  Number of nodes: {len(cluster_indices)}")
        print(f"  Dominant class: {dominant_label} ({dominant_count} nodes, {dominant_count/len(cluster_indices)*100:.1f}%)")
        print(f"  Class distribution:")
        for label, count in zip(unique_true_labels, counts):
            print(f"    Class {label}: {count} nodes ({count/len(cluster_indices)*100:.1f}%)")
    purity = total_correct / total_nodes
    print(f"\n" + "-"*50)
    print(f"OVERALL CLUSTER PURITY: {purity:.4f}")
    print("-"*50)
    return purity

#############################################
#  PIPELINE : Intégration HGT + DEC         #
#############################################
def run_hgt_clustering_dec(g, node_features, relation_embeddings, true_labels, id_node_map, rdf_graph):
    print("\n" + "="*60)
    print("HGT + DEC CLUSTERING PIPELINE")
    print("="*60)

    # Ensure node_features is a dictionary
    if not isinstance(node_features, dict):
        node_features = {i: feat for i, feat in enumerate(node_features)}

    print("\n[STEP 1] Training HGT model...")
    model, embeddings = train_hgt(g, node_features, relation_embeddings)
    if 'author' not in embeddings:
        raise ValueError("No author embeddings found in model output")

    print("\n[STEP 2] Preparing embeddings for clustering...")
    embeddings_dict = {}
    true_labels_dict = {}
    n_clusters_dict = {}

    # Authors
    print("  - Processing author embeddings...")
    author_embeddings = embeddings['author']
    embeddings_dict['author'] = author_embeddings
    true_labels_dict['author'] = true_labels
    n_clusters_dict['author'] = len(np.unique(true_labels))
    print(f"    Authors: {len(author_embeddings)} nodes, {n_clusters_dict['author']} clusters")

    # Publications
    if 'publication' in embeddings:
        print("  - Processing publication embeddings...")
        pub_embeddings = embeddings['publication']
        embeddings_dict['publication'] = pub_embeddings
        pub_true_labels = None
        if hasattr(g.nodes['publication'], 'data') and 'true_domain' in g.nodes['publication'].data:
            pub_true_labels = g.nodes['publication'].data['true_domain'].cpu().numpy()
        if pub_true_labels is None:
            pub_true_labels = get_node_domains(g, 'publication', id_node_map, rdf_graph)
        if pub_true_labels is not None and len(pub_true_labels)>0:
            true_labels_dict['publication'] = pub_true_labels
            n_clusters_dict['publication'] = len(np.unique(pub_true_labels))
            print(f"    Publications: {len(pub_embeddings)} nodes, {n_clusters_dict['publication']} clusters")

    # Conferences
    if 'conference' in embeddings:
        print("  - Processing conference embeddings...")
        conf_embeddings = embeddings['conference']
        embeddings_dict['conference'] = conf_embeddings
        conf_true_labels = None
        if hasattr(g.nodes['conference'], 'data') and 'true_domain' in g.nodes['conference'].data:
            conf_true_labels = g.nodes['conference'].data['true_domain'].cpu().numpy()
        if conf_true_labels is None:
            conf_true_labels = get_node_domains(g, 'conference', id_node_map, rdf_graph)
        if conf_true_labels is not None and len(conf_true_labels)>0:
            true_labels_dict['conference'] = conf_true_labels
            n_clusters_dict['conference'] = len(np.unique(conf_true_labels))
            print(f"    Conferences: {len(conf_embeddings)} nodes, {n_clusters_dict['conference']} clusters")

    # Venues (domains)
    if 'domain' in embeddings:
        print("  - Processing venue embeddings...")
        venue_embeddings = embeddings['domain']
        embeddings_dict['venue'] = venue_embeddings
        # Get proper venue clustering labels based on domain relationships
        venue_true_labels = get_venue_clustering_labels(g, id_node_map, rdf_graph)
        if venue_true_labels is not None and len(venue_true_labels) > 0:
            true_labels_dict['venue'] = venue_true_labels
            n_clusters_dict['venue'] = len(np.unique(venue_true_labels))
            print(f"    Venues: {len(venue_true_labels)} nodes, {len(np.unique(venue_true_labels))} clusters")
        else:
            print("    No proper venue clustering labels found, skipping venue clustering")

    print(f"\n[STEP 3] Running Joint DEC clustering on {len(embeddings_dict)} node types...")
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    dec_model, cluster_assignments = train_joint_dec(embeddings_dict=embeddings_dict,
                                                      n_clusters_dict=n_clusters_dict,
                                                      true_labels_dict=true_labels_dict,
                                                      max_iters=20,
                                                      update_interval=10,
                                                      device=device)

    print("\n[STEP 4] Evaluating clustering results...")
    results_dict = evaluate_joint_dec_clustering(cluster_assignments, true_labels_dict)

    print("\n[STEP 5] Generating visualizations...")

    # Author clustering visualization and analysis
    print("\n  - Visualizing author clusters...")
    visualize_clusters(embeddings_dict['author'], true_labels_dict['author'], cluster_assignments['author'],
                       "Author Clustering Results - HGT with DEC")
    print("\n  - Analyzing author clusters...")
    purity = analyze_clusters(embeddings_dict['author'], true_labels_dict['author'], cluster_assignments['author'],
                              id_node_map)

    # Publication clustering
    if 'publication' in cluster_assignments and 'publication' in true_labels_dict:
        print("\n  - Visualizing publication clusters...")
        visualize_clusters(embeddings_dict['publication'], true_labels_dict['publication'], cluster_assignments['publication'],
                           "Publication Clustering Results - HGT with DEC")

    # Conference clustering
    if 'conference' in cluster_assignments and 'conference' in true_labels_dict:
        print("\n  - Visualizing conference clusters...")
        visualize_clusters(embeddings_dict['conference'], true_labels_dict['conference'], cluster_assignments['conference'],
                           "Conference Clustering Results - HGT with DEC")

    # Venue clustering
    if 'venue' in cluster_assignments and 'venue' in true_labels_dict:
        print("\n  - Visualizing venue clusters...")
        visualize_clusters(embeddings_dict['venue'], true_labels_dict['venue'], cluster_assignments['venue'],
                           "Venue Clustering Results - HGT with DEC")

    print("\n[STEP 6] Saving models and results...")
    # Create output directory
    output_dir = "Hgt with DEC"
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Save models
    torch.save(model.state_dict(), os.path.join(output_dir, 'hgt_model.pt'))
    torch.save(dec_model.state_dict(), os.path.join(output_dir, 'dec_model.pt'))

    # Save embeddings and clusters
    with open(os.path.join(output_dir, 'embeddings_and_clusters.pkl'), 'wb') as f:
        pickle.dump({'embeddings': embeddings_dict, 'clusters': cluster_assignments}, f)

    # Save clustering results summary
    save_clustering_results(results_dict, cluster_assignments, true_labels_dict, output_dir)

    print(f"\n" + "="*60)
    print("PIPELINE COMPLETED SUCCESSFULLY!")
    print(f"All results saved to: {output_dir}/")
    print("="*60)

    return model, embeddings, cluster_assignments

#############################################
#         FONCTIONS D'UTILITÉ RDF         #
#############################################
rdf_type_map = {
    "http://xmlns.com/foaf/0.1/Person": "author",
    "http://example.org/Publication": "publication",
    "http://example.org/Domain": "domain",
    "http://example.org/Conference": "conference",
    "http://example.org/Journal": "journal",
    "http://example.org/Venue": "venue"
}

def get_node_domains(g, node_type, id_node_map, rdf_graph):
    domain_vocab = {}
    domain_counter = 0
    for s, p, o in rdf_graph:
        if str(p) in ["http://example.org/hasDomain", "http://example.org/publishesDomain"]:
            uri = str(o)
            if uri not in domain_vocab:
                domain_vocab[uri] = domain_counter
                domain_counter += 1
    num_nodes = g.number_of_nodes(node_type)
    labels = np.zeros(num_nodes)
    has_domain = False
    if node_type == 'publication':
        domain_relations = ["http://example.org/hasDomain"]
    elif node_type == 'conference':
        domain_relations = ["http://example.org/publishesDomain"]
    else:
        domain_relations = ["http://example.org/hasDomain"]
    for i in range(num_nodes):
        node_uri = id_node_map[int(g.nodes(node_type)[i])]
        for s, p, o in rdf_graph:
            if str(s)==node_uri and str(p) in domain_relations:
                domain_uri = str(o)
                if domain_uri in domain_vocab:
                    labels[i] = domain_vocab[domain_uri]
                    has_domain = True
                    break
    return labels if has_domain else None

def get_venue_clustering_labels(g, id_node_map, rdf_graph):
    """
    Create meaningful venue clustering labels based on domain hierarchies or semantic similarity.
    Instead of giving each venue a unique label, group them into meaningful clusters.
    """
    if 'domain' not in g.ntypes:
        return None

    num_venues = g.number_of_nodes('domain')
    if num_venues == 0:
        return None

    # Strategy 1: Group venues by semantic similarity of their names
    venue_names = []
    venue_uris = []

    for i in range(num_venues):
        venue_uri = id_node_map[int(g.nodes('domain')[i])]
        venue_uris.append(venue_uri)
        # Extract meaningful name from URI
        venue_name = venue_uri.split('/')[-1].lower()
        venue_names.append(venue_name)

    # Create clusters based on semantic similarity
    labels = np.zeros(num_venues, dtype=int)

    # Define domain categories based on common academic domains
    domain_categories = {
        'computer': 0, 'cs': 0, 'software': 0, 'ai': 0, 'ml': 0, 'data': 0,
        'math': 1, 'mathematics': 1, 'statistics': 1, 'algebra': 1,
        'physics': 2, 'quantum': 2, 'mechanics': 2,
        'biology': 3, 'bio': 3, 'medical': 3, 'health': 3,
        'engineering': 4, 'mechanical': 4, 'electrical': 4,
        'chemistry': 5, 'chemical': 5,
        'economics': 6, 'business': 6, 'finance': 6,
        'social': 7, 'psychology': 7, 'sociology': 7
    }

    # Assign labels based on name matching
    for i, name in enumerate(venue_names):
        assigned = False
        for keyword, category in domain_categories.items():
            if keyword in name:
                labels[i] = category
                assigned = True
                break

        if not assigned:
            # If no match found, use a hash-based approach to create consistent groupings
            # This ensures venues with similar names get similar labels
            labels[i] = hash(name) % 5  # Create 5 additional categories

    # Ensure we have at least 2 different labels for meaningful clustering
    unique_labels = len(np.unique(labels))
    if unique_labels < 2:
        # If all venues got the same label, create artificial but meaningful separation
        # Group venues into 3-5 clusters based on their position in the sorted list
        n_clusters = min(5, max(3, num_venues // 2))
        cluster_size = num_venues // n_clusters
        for i in range(num_venues):
            labels[i] = i // cluster_size if i // cluster_size < n_clusters else n_clusters - 1

    print(f"Created {len(np.unique(labels))} venue clusters from {num_venues} venues")
    print(f"Venue cluster distribution: {Counter(labels)}")

    return labels

def get_true_labels(g, rdf_graph, id_node_map):
    category_vocab = {}
    category_counter = 0
    for s, p, o in rdf_graph:
        if str(p) == "http://example.org/hasDomain":
            domain_uri = str(o)
            if domain_uri not in category_vocab:
                category_vocab[domain_uri] = category_counter
                category_counter += 1
    num_authors = g.number_of_nodes('author')
    labels = np.zeros(num_authors)
    for i in range(num_authors):
        author_uri = id_node_map[int(g.nodes('author')[i])]
        for s, p, o in rdf_graph:
            if str(s)==author_uri and str(p)=="http://example.org/hasDomain":
                domain_uri = str(o)
                if domain_uri in category_vocab:
                    labels[i] = category_vocab[domain_uri]
                    break
    return labels, category_vocab

#############################################
#            MAIN – Execution             #
#############################################
if __name__ == "__main__":
    print("\n" + "="*70)
    print("HGT + DEC CLUSTERING PIPELINE - INITIALIZATION")
    print("="*70)

    print("\n[INIT] Loading RDF graph and embeddings...")
    rdf_graph = Graph()
    rdf_graph.parse("DBLP_petit.rdf")

    # Load embeddings
    print("  - Loading node features...")
    with open("node_features.pkl", "rb") as f:
        node_features = pickle.load(f)

    print("  - Loading relation embeddings...")
    with open("relation_embeddings.pkl", "rb") as f:
        relation_embeddings = pickle.load(f)

    print("\n[INIT] Processing RDF graph structure...")
    # RDF type detection
    node_types = {}
    for s in rdf_graph.subjects():
        for p, o in rdf_graph.predicate_objects(subject=s):
            if 'type' in str(p).lower():
                mapped = rdf_type_map.get(str(o))
                if mapped:
                    node_types[str(s)] = mapped
                break

    # Group edges by relation type
    edge_dict = defaultdict(lambda: ([], []))
    class_uri_to_type = {"http://example.org/Conference": "conference", "http://example.org/Journal": "journal", "http://example.org/Venue": "venue"}
    for s, p, o in rdf_graph:
        if isinstance(s, URIRef) and isinstance(o, URIRef):
            if "rdf-syntax-ns#type" in str(p):
                continue
            s_uri, p_uri, o_uri = str(s), str(p), str(o)
            if p_uri.endswith("subClassOf"):
                s_type = class_uri_to_type.get(s_uri, "venue")
                o_type = class_uri_to_type.get(o_uri, "venue")
                rel = "subClassOf"
                edge_dict[(s_type, rel, o_type)][0].append(s_uri)
                edge_dict[(s_type, rel, o_type)][1].append(o_uri)
                continue
            s_type = node_types.get(s_uri, "venue")
            o_type = node_types.get(o_uri, "venue")
            rel = p_uri.split("/")[-1]
            edge_dict[(s_type, rel, o_type)][0].append(s_uri)
            edge_dict[(s_type, rel, o_type)][1].append(o_uri)

    print("\n[INIT] Building DGL heterogeneous graph...")
    all_node_ids = set()
    for (srcs, dsts) in edge_dict.values():
        all_node_ids.update(srcs)
        all_node_ids.update(dsts)
    node_id_map = {uri: idx for idx, uri in enumerate(sorted(all_node_ids))}
    id_node_map = {v: k for k, v in node_id_map.items()}
    data_dict = {}
    for (srctype, reltype, dsttype), (srcs, dsts) in edge_dict.items():
        src_idx = [node_id_map[s] for s in srcs]
        dst_idx = [node_id_map[d] for d in dsts]
        data_dict[(srctype, reltype, dsttype)] = (src_idx, dst_idx)
    g = dgl.heterograph(data_dict)

    # Prepare node features
    features = torch.zeros((len(node_id_map), 768))
    for uri, idx in node_id_map.items():
        if uri in node_features:
            features[idx] = node_features[uri]
    ndata_feats = {}
    for ntype in g.ntypes:
        nids = g.nodes(ntype)
        feats = []
        for nid in nids:
            uri = id_node_map[int(nid)]
            feat = node_features.get(uri, torch.zeros(768))
            feats.append(feat)
        ndata_feats[ntype] = torch.stack(feats)
    g.ndata['feat'] = ndata_feats

    print(f"\n[GRAPH INFO] Constructed DGL heterogeneous graph:")
    print(f"  - Total nodes: {g.num_nodes()}")
    print(f"  - Total edges: {g.num_edges()}")
    print(f"  - Node types: {g.ntypes}")
    print(f"  - Relation types: {list(g.canonical_etypes)}")

    print("\n[INIT] Extracting ground truth labels...")
    true_labels, domain_vocab = get_true_labels(g, rdf_graph, id_node_map)
    print(f"  - Number of domains: {len(domain_vocab)}")
    print(f"  - Author label distribution: {Counter(true_labels)}")
    print("  - Domain mapping:")
    for domain, label in domain_vocab.items():
        print(f"    {domain.split('/')[-1]} -> Label {label}")

    print("\n[INIT] Preparing relation embeddings for HGT...")
    relation_uri_map = {
        'creator': 'http://purl.org/dc/elements/1.1/creator',
        'subClassOf': 'http://www.w3.org/2000/01/rdf-schema#subClassOf',
        'hasDomain': 'http://example.org/hasDomain',
        'publishesDomain': 'http://example.org/publishesDomain',
        'isPartOf': 'http://purl.org/dc/elements/1.1/isPartOf'
    }
    hgt_relation_embeddings = {}
    for etype in g.etypes:
        if etype in relation_uri_map and relation_uri_map[etype] in relation_embeddings:
            hgt_relation_embeddings[etype] = relation_embeddings[relation_uri_map[etype]]
            print(f"  ✓ Found embedding for relation '{etype}'")
        else:
            print(f"  - Creating embedding for relation '{etype}'...")
            from sentence_transformers import SentenceTransformer
            text_model = SentenceTransformer('all-MiniLM-L6-v2')
            embedding = text_model.encode(etype, convert_to_tensor=True)
            if embedding.shape[0]!=768:
                embedding = F.pad(embedding, (0,768-embedding.shape[0]))
            hgt_relation_embeddings[etype] = embedding

    print(f"\n[READY] Starting main pipeline...")
    # Execute the complete pipeline with DEC clustering
    run_hgt_clustering_dec(g, node_features, hgt_relation_embeddings, true_labels, id_node_map, rdf_graph)
```
